Question 1 :Write a Python Program to Check if a Number is Positive, Negative or Zero?

```python
number = int(input("Enter the number to check is Positive, Negative or Zero: "))
def checkNumber(number):
  if(number == 0):
    print("Number is zero ")
  elif(number < 0):
    print(number ,"is negative number ")
  else:
    print(number , "is positive number ")
checkNumber(number)    
```

Question 2: Write a Python Program to Check if a Number is Odd or Even?

```python
number = int(input("Enter the number is even or odd : "))
def EvenOddNumber(number):
  if(number%2 == 0):
    print("Number is even")
  else:
    print(number , "is odd number")
EvenOddNumber(number)      
```

Question 3 : Write a Python to Program to Check Leap Year?

```python
year = int(input("Enter the year to check leap year or not : "))
def CheckLeapYear(year):
  # to check Leap year for century year i.e 2000,1600
  if(year %400 == 0 and year% 100 == 0):
     print(year, "leap year")
  # to check Leap year for non century year i.e 2004,2016, 2020
  elif(year % 4 == 0 and year % 100 != 0):
    print(year,"leap year")
  else:
    print(year,"not a leap year")
CheckLeapYear(year)    
```

Question 4 : Write a Python Program to Check Prime Number?

```python
number =int(input("Enter the number to check prime or not : "))

def checkPrime(number):
  if number>1:
    for i in range(2,number):
        if(number%i == 0):
           print(number, "is not a prime number")
           break
    else:
        print(number, "is prime number")
  else:
     print(number, "is not a prime number")
checkPrime(number)                  
```
